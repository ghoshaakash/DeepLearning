## Get SMILES from SDF

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import SDMolSupplier
from rdkit.Chem import inchi

def extract_name_and_smiles(sdf_file):
    suppl = SDMolSupplier(sdf_file)
    
    for mol in suppl:
        if mol is not None:
            name = mol.GetProp('_Name')
            smiles = Chem.MolToSmiles(mol)
            yield name, smiles

# Example usage
sdf_file_path = 'tox21.sdf'

key_dict={}


for name, smiles in extract_name_and_smiles(sdf_file_path):
    print(f"Name: {name}, SMILES: {smiles}")
    key_dict[name]=smiles
```

```python
import pandas as pd
```

```python
key_df=pd.read_csv("tox21_compoundData.csv")
key_df
```

```python
key_df=key_df.iloc[:,[0,1,2]]
key_df
```

```python
def get_smile(id):
    try:
        return key_dict[id]
    except:
        return None
```

```python
key_df["SMILES"]=key_df["sdftitle"].apply(get_smile)
```

```python
key_df
```

```python
key_df.to_csv("Mol_ID.csv")
```

```python
id_list=key_df["ID"].tolist()
smiles_list=key_df["SMILES"].tolist()
```

```python
smile_id_dict={}
l=len(id_list)
i=0
while(i<l):
    smile_id_dict[id_list[i]]=smiles_list[i]
    i=i+1
```

```python
smile_id_dict
```

```python
import pickle
```

```python
with open('ID_SMILES_dict.pickle', 'wb') as file:
    pickle.dump(smile_id_dict, file)
```

## Playing around with a SMILE value to get a feel for data struct

```python
import matplotlib.pyplot as plt
from pysmiles import read_smiles
import networkx as nx
```

```python
smiles = 'Cl.Nc1cc(C(F)(F)F)ccc1S'
mol = read_smiles(smiles)
    
# atom vector (C only)
print(mol.nodes(data='element'))
# adjacency matrix
print(nx.to_numpy_matrix(mol))

elements = nx.get_node_attributes(mol, name = "element")
nx.draw(mol, with_labels=True, labels = elements, pos=nx.spring_layout(mol))
plt.gca().set_aspect('equal')
```
